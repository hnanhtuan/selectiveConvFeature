```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
font1 = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 25}

font = {'family' : 'normal',
        'size'   : 22}

matplotlib.rc('font', **font1)
matplotlib.rc('font', weight='bold')

matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
```

```python
oxford5k = np.array([0.730293, 0.50, 0.296])
paris6k = np.array([0.768538, 0.5, 0.293532])
holidays = np.array([0.725341, 0.5, 0.257422])
ind = np.arange(3)

oxford5k = oxford5k*100
paris6k = paris6k*100
holidays = holidays*100
width = 0.3

# fig = plt.figure(figsize=(8, 6)) 
fig, ax = plt.subplots(figsize=(8, 6))

ax.grid(True, which='major', linestyle='--', zorder=0)
rects1 = ax.bar(ind, oxford5k, width, color='r', zorder=3)
rects2 = ax.bar(ind + width, paris6k, width, color='g', zorder=3)
rects3 = ax.bar(ind + 2*width, holidays, width, color='b', zorder=3)

ax.set_ylabel(r'$Percentage (\%)$')
# ax.set_title('Scores by group and gender')
ax.set_ylim([0, 100])
ax.set_xticks(ind + width)
ax.set_xticklabels((r'$\mathcal{M}_\textrm{SIFT}$', \
                    r'$\mathcal{M}_\textrm{SUM}$', \
                    r'$\mathcal{M}_\textrm{MAX}$'))

# ax.grid(True, which='both', linestyle='-')

ax.legend((rects1[0], rects2[0], rects3[0]), ('Oxford5k', 'Paris6k', 'Holidays'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%.2f' % np.around(height, 2),
                ha='center', va='bottom', fontsize=20)
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
plt.show()
savefig('test.pdf')
fig.savefig('test.jpg')
```
